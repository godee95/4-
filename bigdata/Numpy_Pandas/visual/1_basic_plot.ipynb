```python
X_1 = range(100) 
Y_1 = [value*3 for value in X_1]

X_2 = range(100) 
Y_2 = [value*2 for value in X_2]

# #eeefff : 흰색
# r: red, 빨간색
plt.plot(X_1, Y_1, color="c")
plt.plot(X_2, Y_2, color="r")

plt.rcParams["figure.figsize"] = (5,3)
plt.show()
```

```python
# linstyle 선 스타일 dashed/dotted
# 그래프 이름 지정 plt.title

plt.plot(X_1, Y_1, color = 'b', linestyle='dashed', marker="x")
plt.plot(X_2, Y_2, color = 'r', linestyle='dotted')

plt.title("Two lines")
plt.show()
```

```python
# label은 범례 표시
# legend(shadow : 범례의 그림자 / fancybox : 범례 박스)
plt.plot(X_1, Y_1, color='b', linestyle='dashed', label='line_1')
plt.plot(X_2, Y_2, color='r', linestyle='dotted', label='line_2')

plt.legend(shadow=True, fancybox=True, loc="lower right")

#plt.show()
```

```python
plt.grid(True, lw=0.4, ls='--', c="r")
plt.plot(X_1, Y_1, color='b', linestyle='dashed', label='line_1')
plt.plot(X_2, Y_2, color='r', linestyle='dotted', label='line_2')

# x,y축의 범위 한계
plt.xlim(-100,200)
plt.ylim(-500,500)

# 라텍스 문자$ , 그래프 제목에 수식을 표현해 주고 싶을 때
# $y=\\frac{분자}{분모}$
plt.title('$y=\\frac{ax+b}{test}$')
plt.show()
```

```python
X_1 = range(100) 
Y_1 = [value*8+6 for value in X_1]

X_2 = range(100) 
Y_2 = [value*2-50 for value in X_2]


plt.plot(X_1, Y_1, color="b", linestyle="dashed") 
plt.plot(X_2, Y_2, color="r", linestyle="dotted")

# 그냥 텍스트를 좌표 상에 올리는 식
plt.text(50, 70, "Line_2222") 

# annotate 주석을 달다
# 덱스트를 화살표와 함께 넘겨주는 방식
# shrink는 화살표 길이?
# xy=(50,100) 화살표 방향 위치
# xytext=(30,110) 화살표 이름 위치
plt.annotate( 'line_2', xy=(50, 100), xytext=(30, 110), arrowprops=dict(facecolor='c', shrink=0.1))
plt.title('$y = ax+b$') 
plt.xlabel('$x_line$') 
plt.ylabel('y_line')
plt.show()
```

```python
# marker 는 표시 점
# marker x면 x표시
# marker ^면 삼각형 표시

# 분포도 보고싶을 때 scatter

# shape 512,3 총 갯수는 512*3
data_1 = np.random.rand(512,3)
data_2 = np.random.rand(512,3)

plt.scatter(data_1[:,0], data_1[:,1], c="b", marker="x")
plt.scatter(data_2[:,0], data_2[:,1], c="r", marker="^")

plt.show()
```

```python
# area 원의 크기 지정
# alpha 투명도 alpha 1에 가까울수록 진해짐

N = 50
x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(N)
area = np.pi * (15*np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha = 0.3)

plt.show()
```

```python
data =[[5., 25., 50., 20.],
      [4., 23., 51., 17],
      [6., 22., 52., 19]]

x = np.arange(4) # 0,1,2,3

# x축은 0,1,2,3, 파란차트 값은 5,25,50,20
plt.bar(x+0.00, data[0], color = "b", width = 0.25)

# x축 기준
# 초록 막대차트는 파란 막대 차트에서 0.25오른쪽으로 이동
# -0.25는 왼쪽으로 이동

# x축은 0.25,1.25,2.25,3.25, 파란차트 값은 4,23,51,17
plt.bar(x+0.25, data[1], color = "g", width = 0.25)
plt.bar(x+0.50, data[2], color = "r", width = 0.25)

# xticks 는 x축의 범례
# x축 0.25,1.25,2.25,3.25 에 "a","b","c","d" 찍어라
plt.xticks(x+0.25,("A","B","C","D"))
plt.show()
```

```python
data = np.array([[5., 25., 50., 20.], 
                 [4., 23., 51., 17], 
                 [6., 22., 52., 19]])
color_list = ['b', 'g', 'r'] 
data_label = ["A","B","C"] 
X = np.arange(data.shape[1])
print(X)

# bottom이라는 옵션은 앞에 있었던 y값의 sum위에다가 그려라.
for i in range(3):
    plt.bar(X, data[i], bottom = np.sum(data[:i], axis=0), color = color_list[i], label=data_label[i])

plt.legend()
plt.show()
```

```python
A = [5., 30., 45., 22.] 
B = [5, 25, 50, 20]
X = range(4)
plt.bar(X, A, color = 'b') 

# bottom=60 위부터
plt.bar(X, B, color = 'r', bottom = 50) 
plt.show()
```

```python
women_pop = np.array([5, 30, 45, 22]) 
men_pop = np.array([5, 25, 50, 20]) 
X = np.arange(4)

plt.barh(X, women_pop, color = 'r') 
# - 있으니깐 왼쪽으로 그려짐
plt.barh(X, -men_pop, color = 'b') 
plt.show()
```
