```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline
np.random.seed(0) 
x = np.random.randn(100)
```

```python
# 데이터 위치를 x축 위에 작은 선분(rug)으로 나타내
# 실제 데이터들의 위치를 보여준다.
sns.rugplot(x);
```

```python
# kernel density 커널 밀도
# 커널이라는 함수를 겹치는 방법으로
# 히스토그램보다 부드러운 형태의 분포 곡선을 보여줌

sns.kdeplot(x);
```

```python
# 러그와 커널 밀도 표시 기능이 있어서
# matplotlib의 hist 명령보다 많이 사용된다.

sns.distplot(x, kde=True, rug=True)
```

# 희귀분석

* 다차원 데이터
* 데이터가 2차원이고 모두 연속적인 실수값이라면 scatter plot을 사용
* scatter plot을 그리기 위해 seaborn 패키지의 jointplot명령을 사용한다.

```python
# kind : 차트의 종류
# kind = 'scatter' : 스캐터 플롯
# kind = 'kde' 커널 밀도 히스토그램

iris = sns.load_dataset("iris") 
sns.jointplot("sepal_width", "petal_length", data=iris, kind="kde", space=0, color="g")
```

```python
sns.regplot(x="total_bill", y="tip", data=tips);
```

```python
# 만약 카테고리형 데이터가 섞여 있는 경우 
# hue 인수에 카테고리 변수 이름 지정
# 카테고리 값에 따라 색상을 다르게 할 수 있다.

sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips); 
```

```python
sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips)
```

```python
# barplot은 카테고리 값에 따른
# 실수 값의 평균과 편차를 표시하는 기본적인 차트
# 평균은 막대의 높이로, 편차는 에러바(error bar)로 표시

sns.barplot(x="day", y="total_bill",hue='sex' , data=tips); 
```

```python
# 각 카테고리 값별로 데이터가 얼마나 있는지 표시할 수 있다.
# 예를 들어 타이타닉호의 각 클래스별, 승객 수
# countplot명령은 데이터 프레임에만 사용할 수 있다.

titanic = sns.load_dataset("titanic") 

# hue 기준으로 좀더 세분화
sns.countplot(x="class", hue="who", data=titanic); 
```

# 2차원 복합 데이터

```python
# 요일별 total_bill
# 박스 하단 하위 25%범위(1.5배, 데이터 있는 범위까지만 표현)
# 박스 하단 상위 75%범위(1.5배, 데이터 있는 범위까지만 표현)
# 박스안에 가로선은 평균값
# 점은 이상치(아웃 라이어)


# boxplot 중앙값, 표준 편차 등, 분포의 간략한 특성 보여줌
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips);
```

```python
# 증감을 나타낼 수 있다.

sns.pointplot(x="time", y="total_bill", hue="smoker", data=tips, dodge=True);
```

* violinplot, stripplot, swarmplot 등은 카테고리값에 따른 각 분포의 실제 데이터나 전체 형상을 보여준다는 장점이 있다.

* stripplot, violinplot, swarmplot 등 에서는 split 또는 dodge 옵션으로 시각화 방법을 변경

```python
# 위에랑 똑같은데, 모양이 다른거 violinplot
# 세로 방향으로 커널 밀도 히스토그램을 그려줌
# 왼쪽과 오른쪽이 대칭이 되도록 하여 바이올린처럼 보인다.
sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, palette="muted")
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, palette="Set2", split=True, scale="count", inner="quartile")
```

```python
# 마치 스캐터 플롯처럼 모든 데이터를 점으로 그려준다.
# jitter=True를 설정하면 가로축상의 위를 무작위로 바꾸어
# 데이터의 수가 많을 경우에 겹치지 않도록 한다.

sns.stripplot(x="day", y="total_bill", hue="smoker", data=tips, jitter=True, palette="Set2", split=True);
```

```python
sns.boxplot(x="tip", y="day", data=tips, whis=np.inf)
```

```python
sns.stripplot(x="tip", y="day", data=tips, jitter=True, color=".3");
```

```python
# stripplot과 비슷하지만 
#데이터를 나타내는 점이 겹치지 않도록 옆으로 이동한다.

sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips); 
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, inner=None) 
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, color="white", edgecolor="gray");
```

```python
# heatmap을 이용해도 
# 두 개의 카테고리 값에 의한 실수 값 변화를 볼 수 있다.

flights = sns.load_dataset("flights") 
flights = flights.pivot("month", "year", "passengers") 
sns.heatmap(flights, annot=True, fmt="d")
```
